```python
import os
import pandas as pd
```

```python
directory = '../Data/Netflix watch log'
```

```python
#for resource ID and timestamp
df_resource = pd.read_csv(os.path.join(directory, "All_Clickstream.csv"))
df_resource["Click Utc Ts"]
```

```python
df_resource["Navigation Level"]
```

```python
df_resource["Navigation Level"] = df_resource["Navigation Level"].apply(lambda x: 'bucket/' + str(x))
df_resource
```

```python
df_resource["Navigation Level"].count()
```

```python
df_timestamp = df_resource[(df_resource["Click Utc Ts"] >= '2023-01-01 00:00:00') & (df_resource["Click Utc Ts"] <= '2023-03-30 00:00:00')]
```

```python
df_timestamp
```

```python
final_df = df_timestamp[["Click Utc Ts"]]
final_df["Resource ID"] = df_timestamp[["Navigation Level"]]
```

```python
final_df
```

```python
df_name = pd.read_csv(os.path.join(directory, "Airline Dataset.csv"))
df_name["Full name"] = df_name["First Name"] + " " + df_name["Last Name"]
```

```python
final_df["Name"] = df_name[["Full name"]]
```

```python
df_action = pd.read_csv(os.path.join(directory, "All_SearchHistory.csv"))
df_action["Action"]
```

```python
final_df = final_df.reset_index(drop=True)
final_df
```

```python
final_df["Action"] = df_action[["Action"]]
final_df = final_df.dropna()
final_df
```

```python
final_df['Click Utc Ts'] = pd.to_datetime(final_df['Click Utc Ts']).dt.strftime('%d-%m-%Y %H:%M')
final_df
```

```python
final_df["Action"].value_counts()
```

```python
import numpy as np
values = ["CREATE", "READ", "WRITE", "UPDATE", "DELETE"]
```

```python
final_df["Action"] = np.random.choice(values, size=len(final_df))
final_df
```

```python
final_df
```

```python
final_df.to_csv(os.path.join("../Data/", "Input.csv"), index=False, header=False)
```
